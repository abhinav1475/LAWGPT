```python
import re
from datasets import load_dataset, Dataset
from transformers import (
    GPT2Tokenizer,
    GPT2LMHeadModel,
    Trainer,
    TrainingArguments,
    DataCollatorForLanguageModeling,
    EarlyStoppingCallback
)

# Step 1: Load the Dataset
print("Loading dataset...")
ds = load_dataset("jizzu/llama2_indian_law_v2")

# Step 2: Extract Questions and Answers
def extract_qa_pairs(entry):
    """Extract question and answer pairs from the text string."""
    match_human = re.search(r'###Human:\s*(.*?)\s*###Assistant:\s*(.*?)(?=###Human:|$)', entry)
    if match_human:
        question = match_human.group(1).strip()
        answer = match_human.group(2).strip()
        return {"question": question, "answer": answer}
    return None

def clean_dataset(dataset):
    """Clean the text fields in the dataset and extract Q&A pairs."""
    qa_pairs = []
    for entry in dataset:
        qa_pair = extract_qa_pairs(entry['text'])  # Accessing the 'text' field
        if qa_pair:  # Ensure we only add successfully extracted pairs
            qa_pairs.append(qa_pair)
    print(f"Extracted {len(qa_pairs)} Q&A pairs.")  # Debug print
    return qa_pairs

# Clean the training and validation sets
print("Cleaning training data...")
cleaned_train_data = clean_dataset(ds['train'])
print("Cleaning validation data...")
cleaned_validation_data = clean_dataset(ds['validation'])

# Step 3: Prepare the Data for Fine-Tuning
# Use only the first 3000 Q&A pairs for training
train_qa_dataset = Dataset.from_dict({
    "question": [pair["question"] for pair in cleaned_train_data[:3000]],  # Only the first 3000 pairs
    "answer": [pair["answer"] for pair in cleaned_train_data[:3000]]
})

validation_qa_dataset = Dataset.from_dict({
    "question": [pair["question"] for pair in cleaned_validation_data],
    "answer": [pair["answer"] for pair in cleaned_validation_data]
})

# Step 4: Tokenization
print("Loading tokenizer...")
tokenizer = GPT2Tokenizer.from_pretrained('distilgpt2')

# Set the padding token to the end of sequence token
tokenizer.pad_token = tokenizer.eos_token

def tokenize_function(examples):
    print("Tokenizing data...")  # Debug print
    inputs = [f"Question: {q} Answer: {a}" for q, a in zip(examples['question'], examples['answer'])]

    # Use the same input as labels, but shift the input IDs to create labels for training
    model_inputs = tokenizer(
        inputs,
        truncation=True,
        padding='max_length',
        max_length=128,
        return_tensors='pt'  # Return PyTorch tensors
    )

    # Set labels to the input IDs
    model_inputs['labels'] = model_inputs['input_ids'].clone()

    return model_inputs

# Apply tokenization to train and validation datasets
print("Tokenizing training dataset...")
train_tokenized = train_qa_dataset.map(tokenize_function, batched=True)
print("Tokenizing validation dataset...")
validation_tokenized = validation_qa_dataset.map(tokenize_function, batched=True)

# Set format for PyTorch tensors
train_tokenized.set_format(type='torch', columns=['input_ids', 'attention_mask', 'labels'])
validation_tokenized.set_format(type='torch', columns=['input_ids', 'attention_mask', 'labels'])

# Step 5: Fine-Tuning the Model
print("Loading model...")
model = GPT2LMHeadModel.from_pretrained('distilgpt2')

# Set training arguments
training_args = TrainingArguments(
    output_dir='/content/drive/MyDrive/LawGPT/results',  # Save results here
    evaluation_strategy='epoch',  # Evaluate at the end of each epoch
    save_strategy='epoch',  # Save the model at the end of each epoch
    learning_rate=2e-5,
    per_device_train_batch_size=2,
    per_device_eval_batch_size=2,
    num_train_epochs=20,  # Increase the number of epochs to 20
    weight_decay=0.01,
    logging_dir='/content/drive/MyDrive/LawGPT/logs',  # Log directory
    logging_steps=10,  # Log every 10 steps
)

# Define the data collator
data_collator = DataCollatorForLanguageModeling(tokenizer=tokenizer, mlm=False)

# Initialize Trainer
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_tokenized,
    eval_dataset=validation_tokenized,
    data_collator=data_collator,  # Use the data collator
)

# Fine-tune the model with early stopping based on validation loss
print("Starting training...")
trainer.train()

# Step 6: Save the Fine-Tuned Model
print("Saving the fine-tuned model...")
model.save_pretrained("/content/drive/MyDrive/LawGPT/fine-tuned-distilgpt2")
tokenizer.save_pretrained("/content/drive/MyDrive/LawGPT/fine-tuned-distilgpt2")

print("Fine-tuning complete and model saved!")

# Step 7: Print Validation Metrics
print("Evaluating on validation set...")
metrics = trainer.evaluate()
print(f"Validation metrics: {metrics}")
```
